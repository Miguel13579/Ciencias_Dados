```python
# LLM simulada simples (nÃ£o faz chamadas externas)
class SimulatedLLM:
    def chat(self, messages, **kwargs):
        return {"content": "Simulado: " + messages[-1]['content']}
```

```python
from crewai import Agent

simulated_llm = SimulatedLLM()

def criar_agente(nome, objetivo):
    return Agent(
        role=nome,
        goal=objetivo,
        backstory=f"{nome} responsÃ¡vel por {objetivo.lower()}",
        allow_delegation=True,
        llm=simulated_llm  # ðŸ‘ˆ Evita chamadas Ã  OpenAI
    )

gerente = criar_agente("Gerente de Projetos", "interpretar a solicitaÃ§Ã£o e dividir tarefas")
dev = criar_agente("Desenvolvedor TÃ©cnico", "criar scripts conforme solicitado")
analista = criar_agente("Analista de Resultados", "avaliar a saÃ­da e gerar um relatÃ³rio")
```

```python
from crewai import Task, Crew

tarefa1 = Task(
    description="Interpretar a solicitaÃ§Ã£o do usuÃ¡rio",
    expected_output="Lista de subtarefas distribuÃ­das",
    agent=gerente
)

tarefa2 = Task(
    description="Criar cÃ³digo de anÃ¡lise de sentimentos em Python",
    expected_output="Script funcional em Python para anÃ¡lise de sentimentos",
    agent=dev
)

tarefa3 = Task(
    description="Analisar os resultados da execuÃ§Ã£o e gerar relatÃ³rio",
    expected_output="Resumo com interpretaÃ§Ã£o dos resultados e sugestÃµes de melhoria",
    agent=analista
)
```

```python
from textblob import TextBlob

tweets = [
    "AI is amazing!",
    "I hate how AI is invading our lives.",
    "So impressed with ChatGPT.",
    "AI makes me nervous.",
    "I love using AI to automate tasks.",
    "AI will destroy jobs.",
    "Machine learning is fascinating.",
    "AI isn't as smart as people think.",
    "Incredible advances in deep learning.",
    "AI tools are fun to use!"
]

resultados = [(t, TextBlob(t).sentiment.polarity) for t in tweets]

positivos = sum(1 for _, score in resultados if score > 0)
negativos = sum(1 for _, score in resultados if score < 0)
neutros = sum(1 for _, score in resultados if score == 0)

print("\nðŸ“Š RelatÃ³rio de Sentimentos:")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Neutros: {neutros}")
```

```python
# ================================
# ðŸ“„ GERAÃ‡ÃƒO DE RELATÃ“RIO FINAL
# ================================
print("\n" + "="*40)
print("ðŸ“„ RELATÃ“RIO FINAL")
print("="*40)

# 1. SolicitaÃ§Ã£o
print("\nðŸ“Œ SolicitaÃ§Ã£o do UsuÃ¡rio:")
print("Preciso de uma anÃ¡lise de sentimentos de 10 tweets sobre IA")

# 2. SoluÃ§Ã£o Gerada
print("\nðŸ› ï¸ SoluÃ§Ã£o Gerada:")
print("Foi criado um script em Python utilizando a biblioteca TextBlob para realizar a anÃ¡lise de sentimentos de 10 tweets simulados.")

# 3. ExecuÃ§Ã£o da Tarefa
print("\nðŸš€ ExecuÃ§Ã£o da Tarefa:")
for tweet, score in resultados:
    sentimento = "Positivo" if score > 0 else "Negativo" if score < 0 else "Neutro"
    print(f"â†’ \"{tweet}\" => {sentimento} (score={score:.2f})")

print("\nðŸ“Š Resumo da ExecuÃ§Ã£o:")
print(f"Tweets Positivos: {positivos}")
print(f"Tweets Negativos: {negativos}")
print(f"Tweets Neutros: {neutros}")

# 4. Resumo Final
print("\nðŸ§  Resumo Final:")
if positivos > negativos:
    interpretacao = "A maioria dos tweets tem sentimento positivo em relaÃ§Ã£o Ã  IA."
elif negativos > positivos:
    interpretacao = "A maioria dos tweets expressa sentimento negativo em relaÃ§Ã£o Ã  IA."
else:
    interpretacao = "Os tweets estÃ£o balanceados entre sentimentos positivos e negativos."

print(interpretacao)
print("SugestÃ£o: Pode-se investigar mais profundamente o porquÃª de opiniÃµes negativas, caso haja interesse.")
```
