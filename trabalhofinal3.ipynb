```python
# LLM simulada simples (não faz chamadas externas)
class SimulatedLLM:
    def chat(self, messages, **kwargs):
        return {"content": "Simulado: " + messages[-1]['content']}
```

```python
from crewai import Agent

simulated_llm = SimulatedLLM()

def criar_agente(nome, objetivo):
    return Agent(
        role=nome,
        goal=objetivo,
        backstory=f"{nome} responsável por {objetivo.lower()}",
        allow_delegation=True,
        llm=simulated_llm  # 👈 Evita chamadas à OpenAI
    )

gerente = criar_agente("Gerente de Projetos", "interpretar a solicitação e dividir tarefas")
dev = criar_agente("Desenvolvedor Técnico", "criar scripts conforme solicitado")
analista = criar_agente("Analista de Resultados", "avaliar a saída e gerar um relatório")
```

```python
from crewai import Task, Crew

tarefa1 = Task(
    description="Interpretar a solicitação do usuário",
    expected_output="Lista de subtarefas distribuídas",
    agent=gerente
)

tarefa2 = Task(
    description="Criar código de análise de sentimentos em Python",
    expected_output="Script funcional em Python para análise de sentimentos",
    agent=dev
)

tarefa3 = Task(
    description="Analisar os resultados da execução e gerar relatório",
    expected_output="Resumo com interpretação dos resultados e sugestões de melhoria",
    agent=analista
)
```

```python
from textblob import TextBlob

tweets = [
    "AI is amazing!",
    "I hate how AI is invading our lives.",
    "So impressed with ChatGPT.",
    "AI makes me nervous.",
    "I love using AI to automate tasks.",
    "AI will destroy jobs.",
    "Machine learning is fascinating.",
    "AI isn't as smart as people think.",
    "Incredible advances in deep learning.",
    "AI tools are fun to use!"
]

resultados = [(t, TextBlob(t).sentiment.polarity) for t in tweets]

positivos = sum(1 for _, score in resultados if score > 0)
negativos = sum(1 for _, score in resultados if score < 0)
neutros = sum(1 for _, score in resultados if score == 0)

print("\n📊 Relatório de Sentimentos:")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Neutros: {neutros}")
```

```python
# ================================
# 📄 GERAÇÃO DE RELATÓRIO FINAL
# ================================
print("\n" + "="*40)
print("📄 RELATÓRIO FINAL")
print("="*40)

# 1. Solicitação
print("\n📌 Solicitação do Usuário:")
print("Preciso de uma análise de sentimentos de 10 tweets sobre IA")

# 2. Solução Gerada
print("\n🛠️ Solução Gerada:")
print("Foi criado um script em Python utilizando a biblioteca TextBlob para realizar a análise de sentimentos de 10 tweets simulados.")

# 3. Execução da Tarefa
print("\n🚀 Execução da Tarefa:")
for tweet, score in resultados:
    sentimento = "Positivo" if score > 0 else "Negativo" if score < 0 else "Neutro"
    print(f"→ \"{tweet}\" => {sentimento} (score={score:.2f})")

print("\n📊 Resumo da Execução:")
print(f"Tweets Positivos: {positivos}")
print(f"Tweets Negativos: {negativos}")
print(f"Tweets Neutros: {neutros}")

# 4. Resumo Final
print("\n🧠 Resumo Final:")
if positivos > negativos:
    interpretacao = "A maioria dos tweets tem sentimento positivo em relação à IA."
elif negativos > positivos:
    interpretacao = "A maioria dos tweets expressa sentimento negativo em relação à IA."
else:
    interpretacao = "Os tweets estão balanceados entre sentimentos positivos e negativos."

print(interpretacao)
print("Sugestão: Pode-se investigar mais profundamente o porquê de opiniões negativas, caso haja interesse.")
```
